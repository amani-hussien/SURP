```python
import model_universe_2d as u
import matplotlib.pyplot as plt
```

default parameters: $\delta* = 10$ and $p=7$, $R=3$

```python
# density fields - aviles's parameters
rho_x = u.rho_x  # tracer density field 
rho_R = u.rho_R # smoothed density tracer field
delta_x = u.delta_x # overdensity of tracer field
delta_R = u.delta_R # overdensity of downsampled tracer field
m = u.m # mark
delta_M = u.delta_M # marked density field

u.plot_heatmap(rho_x, 'original density field')
```

### testing R values

##### universe 1 

```python
# different R values

rho_R = u.gaussian_filter(rho_x, R=1)
delta_x = u.overdensity_field(rho_x)
delta_R1 = u.overdensity_field(rho_R)
m1, delta_M = u.marked_overdensity_field(delta_x, delta_R1, delta_star=4, p=10)


rho_R = u.gaussian_filter(rho_x, R=2)
delta_x = u.overdensity_field(rho_x)
delta_R2 = u.overdensity_field(rho_R)
m2, delta_M = u.marked_overdensity_field(delta_x, delta_R2, delta_star=4, p=10)


rho_R = u.gaussian_filter(rho_x, R=3)
delta_x = u.overdensity_field(rho_x)
delta_R3 = u.overdensity_field(rho_R)
m3, delta_M = u.marked_overdensity_field(delta_x, delta_R3, delta_star=4, p=10)


rho_R = u.gaussian_filter(rho_x, R=4)
delta_x = u.overdensity_field(rho_x)
delta_R4 = u.overdensity_field(rho_R)
m4, delta_M = u.marked_overdensity_field(delta_x, delta_R4, delta_star=4, p=10)

rho_R = u.gaussian_filter(rho_x, R=5)
delta_x = u.overdensity_field(rho_x)
delta_R5 = u.overdensity_field(rho_R)
m5, delta_M = u.marked_overdensity_field(delta_x, delta_R4, delta_star=4, p=10)

# stacked overdensity plots
fig, ax = plt.subplots(1,5,figsize=(10,8))

ax[0].imshow(delta_x, cmap='plasma') 
ax[0].set_title("original")
ax[1].imshow(delta_R1, cmap='plasma') 
ax[1].set_title("R=1")
ax[2].imshow(delta_R2, cmap='plasma') 
ax[2].set_title("R=2")
ax[3].imshow(delta_R3, cmap='plasma') 
ax[3].set_title("R=3")
ax[3].imshow(delta_R4, cmap='plasma') 
ax[3].set_title("R=4")
ax[4].imshow(delta_R5, cmap='plasma') 
ax[4].set_title("R=5")


plt.tight_layout()
plt.show()


# stacked mark

fig, ax = plt.subplots(1,5,figsize=(10,8))


ax[0].imshow(delta_x, cmap='plasma') 
ax[0].set_title("original")
ax[1].imshow(m1, cmap='plasma') 
ax[1].set_title("R=1")
ax[2].imshow(m2, cmap='plasma') 
ax[2].set_title("R=2")
ax[3].imshow(m3, cmap='plasma') 
ax[3].set_title("R=3")
ax[3].imshow(m4, cmap='plasma') 
ax[3].set_title("R=4")
ax[4].imshow(m5, cmap='plasma') 
ax[4].set_title("R=5")


plt.tight_layout()
plt.show()




```
