```python
import model_universe_3d as u
import plotly.graph_objects as plotly
from plotly.subplots import make_subplots


# density fields
rho_x = u.rho_x  # tracer density field
rho_R = u.rho_R # smoothed density tracer field
delta_x = u.delta_x # overdensity of tracer field
delta_R = u.delta_R # overdensity of downsampled tracer field
m = u.m # mark
delta_M = u.delta_M # marked density field
```

```python
u.plot_isosurface(rho_x, 'tracer density field')
```

```python
u.plot_volume(rho_R, 'smoothed density field')
```

```python
u.plot_isosurface(rho_R, 'smoothed density field')
```

```python
X, Y, Z, rho_x = u.load_universe("universe_parameters.json")

rho_R = u.gaussian_filter(rho_x, R=3)
delta_x = u.overdensity_field(rho_x)
delta_R = u.overdensity_field(rho_R)
m, delta_M = u.marked_overdensity_field(delta_x, delta_R, delta_star=4, p=10)

# u.plot_isosurface(delta_x, 'tracer overdensity field')
# u.plot_isosurface(m, 'mark')
# u.plot_isosurface(delta_M, 'marked density field')
```

```python
# 1. Define a 1x3 grid and specify that all subplots are 3D
fig = make_subplots(
    rows=1,
    cols=3,
    specs=[[{"type": "scene"}, {"type": "scene"}, {"type": "scene"}]],
    subplot_titles=("tracer overdensity", "mark", "marked field"),
)

# 2. Add the first Volume plot to Row 1, Col 1
fig.add_trace(
    plotly.Volume(
        x=X.flatten(),
        y=Y.flatten(),
        z=Z.flatten(),
        value=delta_x.flatten(),
        isomin=0,
        isomax=delta_x.max(),
        opacity=0.08,
        surface_count=60,
        colorbar=dict(len=0.3, y=0.5, x=-0.07),  # Moves colorbar to left
    ),
    row=1,
    col=1,
)

# 3. Add the second Volume plot to Row 1, Col 2
fig.add_trace(
    plotly.Volume(
        x=X.flatten(),
        y=Y.flatten(),
        z=Z.flatten(),
        value=m.flatten(),
        isomin=0,
        isomax=m.max(),
        opacity=0.08,
        surface_count=60,
        colorbar=dict(len=0.3, y=0.5, x=0.45),  # Moves colorbar to middle
    ),
    row=1,
    col=2,
)

# 4. Add the third Volume plot to Row 1, Col 3
fig.add_trace(
    plotly.Volume(
        x=X.flatten(),
        y=Y.flatten(),
        z=Z.flatten(),
        value=delta_M.flatten(),
        isomin=0,
        isomax=delta_M.max(),
        opacity=0.08,
        surface_count=60,
        colorbar=dict(len=0.3, y=0.5, x=1.02),  # Moves colorbar to right
    ),
    row=1,
    col=3,
)

# 5. Apply layout settings
fig.update_layout(
    template="plotly_dark",
    height=600,
    width=1600,  # Give it extra width for side-by-side 3D rendering
    title_text="Comparison",
)

fig.show()
```
