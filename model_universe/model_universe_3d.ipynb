```python
import model_universe_3d as u
import plotly.graph_objects as plotly
import plotly.express as px
from plotly.subplots import make_subplots
import numpy as np
```

```python

# load a universe 

X, Y, Z, rho_x = u.load_universe("universe_parameters.json")

rho_R = u.gaussian_filter(rho_x, R=3)
delta_x = u.overdensity_field(rho_x)
delta_R = u.overdensity_field(rho_R)
m_a, delta_Ma = u.marked_overdensity_field(delta_x, delta_R, delta_star=10, p=7)
m_m, delta_Mm = u.marked_overdensity_field(delta_x, delta_R, delta_star=4, p=9)
residual_m = np.abs(m_a - m_m)
residual_deltaM = np.abs(delta_Ma -delta_Mm)
```

### comparing mark parameter choices 

plot:
- tracer density field 
- aviles' parameters
- my parameters
- the residual

```python
fig = make_subplots(
    rows=1,
    cols=4,
    specs=[[{"type": "scene"}, {"type": "scene"}, {"type": "scene"}, {"type": "scene"}]],
    subplot_titles=("(a) Tracer overdensity", "(b) Marked field: Aviles' parameters", "(c) Marked field: my parameters", "(d) Residual"),
)

fig.add_trace( 
    plotly.Isosurface(
        x=X.flatten(),
        y=Y.flatten(),
        z=Z.flatten(),
        value=delta_x.flatten(),
        isomin=delta_x.min(),
        isomax=delta_x.max(),
        opacity=0.08,
        surface_count=20,
        colorscale="plasma",
        colorbar=dict(len=0.3, y=0.5, x=-0.07), 
    ),
    row=1,
    col=1,
)

fig.add_trace( 
    plotly.Isosurface(
        x=X.flatten(),
        y=Y.flatten(),
        z=Z.flatten(),
        value=delta_Ma.flatten(),
        isomin=delta_Ma.min(),
        isomax=delta_Ma.max(),
        opacity=0.08,
        surface_count=20,
        colorscale="plasma",
        colorbar=dict(len=0.3, y=0.5, x=0.22),  
    ),
    row=1,
    col=2,
)

fig.add_trace(
    plotly.Isosurface(
        x=X.flatten(),
        y=Y.flatten(),
        z=Z.flatten(),
        value=delta_Mm.flatten(),
        isomin=delta_Mm.min(),
        isomax=delta_Mm.max(),
        opacity=0.08,
        surface_count=20,
        colorscale="plasma",
        colorbar=dict(len=0.3, y=0.5, x=0.73),  
    ),
    row=1,
    col=3,
)

fig.add_trace( 
    plotly.Isosurface(
        x=X.flatten(),
        y=Y.flatten(),
        z=Z.flatten(),
        value=residual_deltaM.flatten(),
        isomin=residual_deltaM.min(),
        isomax=residual_deltaM.max(),
        opacity=0.08,
        surface_count=20,
        colorscale="plasma",
        colorbar=dict(len=0.3, y=0.5, x=1.02), 
    ),
    row=1,
    col=4,
)

fig.update_layout(
    template="plotly_dark",
    height=600,
    width=2200,  
    title_text="Comparison of Parameter Choices",
)

fig.write_html("comparison-markf.html")
fig.show(renderer='browser')
```
